```python
#soal2 Buatlah fungsi untuk menghitung jumlah kasus harian Covid-19 (dataset yang disediakan merupakan jumlah kasus kumulatif)!
# buat fungsi kasus harian
# buat kasus hari t - kasus komulatif t -1, karena data nya komulatif. misal kasus Vietnam tanggal 1/22/21 adalah 2, dan 1/23/21 adalah 2. berarti tidak ada penambahan kasus di tanggal 1/23/21 

def daily_case(df):
    df = df.set_index("Country/Region")
    df = df.drop(["Province/State", "Lat", "Long"], axis=1)
    df = df.transpose()
    df = df.diff().fillna(0)
    return df
```

```python
df_daily = daily_case(data)
df_daily.head()
```
